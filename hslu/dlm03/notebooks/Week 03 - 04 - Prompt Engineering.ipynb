```python
import random

import jinja2

from hslu.dlm03.common import backend
```

```python
BACKEND = backend.Gemini2p5Flash().get_backend()
```

# Zero Shot Prompting

```python
ZERO_SHOT_PROMPT = """
!!!Write your Prompt Here!!!
"""
messages = [{"role": "user", "content": ZERO_SHOT_PROMPT}]
response = BACKEND(messages=messages)
choice = random.choice(response.choices)
message = choice.message
print(message.content)
```

# K-Shot Prompting

```python
shots = []
K_SHOT_PROMPT_TEMPLATE_STRING = """
!!!Write your Prompt Template Here!!!
"""
K_SHOT_PROMPT_TEMPLATE = jinja2.Template(K_SHOT_PROMPT_TEMPLATE_STRING)
messages = [{"role": "user", "content": K_SHOT_PROMPT_TEMPLATE.render(shots=shots)}]
response = BACKEND(messages=messages)
choice = random.choice(response.choices)
message = choice.message
print(message.content)
```

# Chain-of-Thought Prompting

```python
COT_PROMPT = """
!!!Write your Prompt Here!!!
"""
messages = [{"role": "user", "content": COT_PROMPT}]
response = BACKEND(messages=messages)
choice = random.choice(response.choices)
message = choice.message
print(message.content)
```

# Self-Consistency

```python
trials = 5
messages = [{"role": "user", "content": COT_PROMPT}]
response = BACKEND(messages=messages, n=trials)
choices = response.choices

SELF_CONSISTENCY_PROMPT_TEMPLATE_STRING = """
!!!Write your Prompt Here!!!
"""
SELF_CONSISTENCY_PROMPT_TEMPLATE = jinja2.Template(SELF_CONSISTENCY_PROMPT_TEMPLATE_STRING)
messages = [{"role": "user", "content": SELF_CONSISTENCY_PROMPT_TEMPLATE.render(choices=choices)}]
response = BACKEND(messages=messages, n=trials)
choice = random.choice(response.choices)
message = choice.message
print(message.content)
```

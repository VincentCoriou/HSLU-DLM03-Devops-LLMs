```python
import os

from hslu.dlm03.common import agent, backend, tools
from hslu.dlm03.util import ipython_utils
```

```python
BACKEND = backend.Gemini2p5Flash().get_async_backend()
```

```python
GITHUB_MCP_SERVER_URL = "https://api.githubcopilot.com/mcp/"
GITHUB_PAT = os.environ["GITHUB_PAT"]
ALLOWED_TOOLS = {"get_file_contents"}
TOOL_MANAGER = tools.ToolManager.from_url(GITHUB_MCP_SERVER_URL, authorization=GITHUB_PAT, allowed_tools=ALLOWED_TOOLS)
```

```python
SYSTEM_INSTRUCTION = """You are an Expert Senior Developer whose purpose is to help more junior developers in your team understand some github repository.
You will be provided with a Github Repository Link to start with, and you should answer the question that the junior developer asks you about it.
You should use your tools to go interact with the repository.
Note that github URLs are of the form https//github.com/{name}/{repository}"""
```

```python
AGENT = agent.Agent(BACKEND, TOOL_MANAGER)
```
